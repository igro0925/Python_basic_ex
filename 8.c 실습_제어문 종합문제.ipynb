```python
print("""
### 1. 고객 등급 분류 (조건문)

고객의 구매 금액에 따라 등급을 나누는 프로그램을 작성하세요.

- 100만 원 이상: **VIP**
- 50만 원 이상: **Gold**
- 10만 원 이상: **Silver**
- 그 외: **Bronze**

👉 입력: 구매 금액
👉 출력: 등급 문자열
""")
```

```python
amount = int(input("구매 금액 입력: "))
```

```python
amount
```

```python
if amount >= 1000000 :
  print("**VIP**")
elif amount >= 500000 :
  print("**Gold**")
elif amount >= 100000 :
  print("**silver**")
else:
  print("**Bronze**")
```

```python
print(
"""
### 3. 재고 소진 알림 (조건문 + 반복문)

상품 재고가 10개 있습니다.

고객이 차례로 주문할 때마다 재고를 차감하고, 재고가 0이 되면 "품절되었습니다"를 출력한 뒤 주문 처리를 중단하세요.

👉 입력: 주문 수량 리스트 `[2, 3, 5, 1]`

👉 출력: 각 주문 처리 결과 + 품절 알림
""")
```

```python
orders = [2,3,5,1]
stock = 10

# 주문 수량을 가져와서, 재고와 비교하여 처리가 가능한지, 불가능한지 판별
for order in orders:
  if stock >= order:
    # 재고가 주문갯수보다 충분하니까 처리가능
    # 1. 재고의 갯수를 업데이트
    stock = stock - order
    # 2. 주문 처리 완료됐다고 출력함
    print(f"주문 처리 완료. 남은 재고는 {stock}개 입니다.")
  else:
    # 재고가 부족해서 처리 불가, 반복문을 끝냄
    break
if stock == 0:
  print("품절입니다.")
```
