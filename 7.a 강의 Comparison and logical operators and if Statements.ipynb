### 비가 오면 우산을 들고 나가고, 비가 오지 않으면 놓고간다.

### 5가 2보다 크면 condition is True 를 출력하고, 그렇지 않으면, condition is False를 출력한다.

```python
# 조건 + 행동(액션) ==> 조건문
```

```python
if 5>2:
  print("조건이 참이면 실행")
```

```python
my_score = 80
cut_line = 60

# 참인 경우와 거짓인 경우 각각에 대해 판단해서 실행

if my_score >= cut_line :
  print("합격")
else :
  print("불합격")
```

```python
# 60 미만 : 불합격
# 60 ~ 80미만 : 합격
# 80이상 : 최고

my_score = 80
if my_score >= 80 :
  print("최고")
elif my_score >=60 and my_score < 80:
  print("합격")
else :
  print("불합격")
```

## Nested 중첩된 조건문

```python
# 메모리 상태 변화로 설명하기
year = 2025
month = 12
day = 31
```

```python
day = day + 1
if day > 31:
  month = month + 1
  day = 1

  if month > 12:
    year = year + 1
    month = 1
print(f"{year}년 {month}월 {day}일")
```
