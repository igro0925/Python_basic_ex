# <font color='blue'> 9. 유저가 점수를 입력하면, 해당 학점을 출력하는 프로그램입니다.  
# <font color='blue'> 단, 유저가 -1 을 입력하면, 프로그램을 종료합니다. -1 을 입력하기 전에는 계속 입력을 받습니다.


- 사용자가 어떤 값을 입력함. 그 값에 따라 A,B,C,F 등급을 표시함
- 숫자 <= 73 :F
- 74 <= 숫자 < 84 : C
- 84 <= 숫자 < 94 : B
- 94 <= 숫자 <= 100 : A

```python
score = 0
while score != -1:
  score = int(input())
  if score <= 73:
    print("F등급입니다.")
  elif score >=73 and score < 84:
    print("C등급입니다.")
  elif score >=84 and score < 94:
    print(f"B등급입니다.")
  elif score >=94 and score < 100:
    print("A등급입니다.")
```

```python
# 업그레이드 1~100이내의 숫자를 적지 않고 그 외의 숫자를 적을때 오류표시
score = 0
while score != -1:
  score = int(input())
  if score <= 73:
    print("F등급입니다.")
  elif score >=73 and score < 84:
    print("C등급입니다.")
  elif score >=84 and score < 94:
    print(f"B등급입니다.")
  elif 94 <= score <= 100:
    print("A등급입니다.")
  else:
    print("1에서 100이내의 숫자를 적어주세요.")
```

```python
# 선생님 풀이
while True:
  score = int(input("숫자 입력 >>> "))
  if score == -1:
    print("종료")
    break
  else:
    if score <= 73:
      print("F등급입니다.")
    elif score >=73 and score < 84:
      print("C등급입니다.")
    elif score >=84 and score < 94:
      print(f"B등급입니다.")
    elif 94 <= score <= 100:
      print("A등급입니다.")
```

# <font color='blue'> 입력)  단을 입력하세요.

```python
num = int(input())
while num != 0:
  if num < 0 or num > 9:
    print("잘못 입력하셨습니다.")
    num = int(input())
    continue
  if num > 0 and num <=9:
    for i in range(num,10):
      for j in range(1,10):
        print(f"{i}x{j}={i*j}")
    break
```

```python
spam_cnt = 0

for  word in spam_words:
  if word in sms:
    spam_cnt = spam_cnt + 1
    break
  else:
    pass

if spam_cnt > 0:
  print("스팸 메세지 입니다.")
  else:
    print("스팸 메세지가 아닙니다.")
```
