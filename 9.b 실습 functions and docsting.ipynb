# <font color='blue'> 1.1. 하나의 숫자를 입력받아서, 입력받은 수까지의 짝수만 리스트로 반환하는 함수를 만드세요.
    예) 7을 입력하면, 결과로 [2,4,6] 리스트를 반환.

```python
def even_num(num):
  i = 1
  return_list = []
  while i <= num:
    if i % 2 == 0:
      return_list.append(i)
      print(return_list)
    i = i+1
  return return_list
```

```python
num = int(input())
list1 = even_num(num)
list1
```

```python
# 정의부
def print_even(n):
  return list(range(0, n+1, 2))
```

```python
# 호출부
# 0~n까지 짝수 출력
n = 20
result = print_even(n)
print(result)
```

# <font color='blue'> 2. 숫자 두개, 문자열 1개를 입력받아서, 더하기와 곱하기 연산을 하고, 리턴하는 함수를 만드세요.
## 함수명은 cal ( value1, value2, operator_str )
## operator_str 은 'add', 'mul' 두가지 입니다.
## 만약 cal(3, 4, 'add') 이면 더하기를 하고
## cal(3, 4, 'mul') 이면 곱하기를 합니다.

```python
def  cal ( value1, value2, operator_str ):
  if operator_str == "add":
    return value1 + value2
  elif operator_str == "mul":
    return value1 * value2
```

```python
value1 = 3
value2 = 4
str1 = input()

print(cal(value1, value2, str1))
```

```python
# 정의부
def  cal ( value1, value2, operator_str ):
  # pass  초반에 함수 안의 값 채우기 용
  # 'add'
  if operator_str == "add" :
    return value1 + value2

  # 'mul'
  elif operator_str == "mul":
    return value1 * value2

  else:
    #  잘못입력 예외 처리
    return "입력값이 잘못됐습니다."

  # 곱하기 연산결과 리턴
```

```python
# 호출부
# value1 = 5
# value2 = 20
# operator_str = 'add'  # add, mul

value1 = int(input("숫자를 입력해 주세요 >> "))
value2 = int(input("숫자를 입력해 주세요 >> "))

operator_str = input("연산 방법 입력('add', 'mul)")

cal ( value1, value2, operator_str )
```

# 3. 과일 등급을 반환하는 함수를 작성하세요.

## 과일 크기를 입력으로 받으면, 등급을 리턴하는 함수.
    
## 1에서10 : C등급, 11에서20 : B등급, 21에서30 : A등급
    

```python
def size_grade(size):
  if 21 <= size <= 30:
    return "A"
  elif 11 <= size <= 20:
    return "B"
  elif 1 <= size <= 10:
    return "C"
```

```python
size = 10
size_grade(size)
```

# <font color='blue'> 4. 다음 리스트가 있다. 이 리스트를 함수에 넘겨주면(파라미터), 리스트의 요소를 모두 더한 값을 리턴하는 함수 arr_sum 함수를 작성하세요.

```python
def arr_sum(a):
  sum = 0
  for i in a:
    sum = sum + i
  return sum
```
