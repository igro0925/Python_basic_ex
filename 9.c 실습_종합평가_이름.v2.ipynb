```python
# - 항목별 연간 총 매출을 계산하여 딕셔너리 형태로 출력하세요.
sales_dict1 = {}

for i in sales_data:
  sum = 0
  for j in sales_data[i]:
    # print(sales_data[i][j])
    sum = sum + sales_data[i][j]
  sales_dict1[i] = sum
print(f"항목별 연간 총 매출 : {sales_dict1}")
```

```python
# - 분기별 총 매출을 계산하여 딕셔너리 형태로 출력하세요.
sales_dict2 = {}

type(sales_data["T-shirts"]["Q1"])
cnt = 0
for i in sales_data:
  for j in sales_data[i]:
    if cnt == 0:
      sales_dict2[j] = sales_data[i][j]
    else:
      sales_dict2[j] = sales_dict2[j] + sales_data[i][j]
  cnt = 1

print(f"분기별 연간 총 매출 : {sales_dict2}")
```

```python
# - 연간 매출이 가장 높은 최고 매출 항목(Best Item) 을 찾으세요.
# 개판이라 죄송합니다ㅠ

cnt = 0
best_item_sales = 0
item = ""

for i in sales_dict1:
  if cnt == 0:
    best_item_sales = sales_dict1[i]
    item = i
  else:
    if best_item_sales < sales_dict1[i]:
        best_item_sales = sales_dict1[i]
        item = i
  cnt = 1

print(f"최고 매출 항목(Best Item)과 매출액 : {item} - {best_item_sales}")
```

```python
"""
## 4. 매출 목표 달성 여부 확인 함수

- 함수 이름: `check_sales(sales, target=1000000)`
- 기능: 월별 매출액(`sales`)을 입력받아 목표(`target`) 이상인지 판별.
- 출력: 목표 달성 여부에 따라 `"목표 달성"` 또는 `"목표 미달"`.
- 활용 예: KPI 평가 자동화.

# 사용 예시

print(check_sales(1200000))  # 목표 달성
print(check_sales(800000))   # 목표 미달
"""
```

```python
def check_sales(sales, target=1000000):
  if sales >= target:
    return "목표달성"
  else:
    return "목표미달"

```

## 5. 고객 등급 분류 함수(10점)

```python
"""
## 5. 고객 등급 분류 함수

- 함수 이름: `classify_customer(amount)`
- 기능: 고객의 구매금액(`amount`)에 따라 등급 분류.
    - 100만 이상 → VIP
    - 50만 이상 → GOLD
    - 10만 이상 → SILVER
    - 그 외 → BRONZE
- 활용 예: 마케팅 타겟팅 분류.

# 사용 예시

print(classify_customer(1200000))  # VIP
print(classify_customer(300000))   # SILVER
"""
```

```python
def classify_customer(amount):
  if amount >= 1000000:
    return "VIP"
  elif amount >=500000:
    return "GOLD"
  elif amount >=100000:
    return "SILVER"
  else:
    return "BRONZE"
```

```python
print(classify_customer(1200000))  # VIP
print(classify_customer(300000))   # SILVER
```

## 6. 업무 자동 보고서용 평균 처리 함수(10점)

```python
"""
## 6. 업무 자동 보고서용 평균 처리 함수

- 함수 이름: `calc_metrics(data, metric="avg")`
- 기능: 숫자 리스트(`data`)를 입력받아 평균(`avg`), 최대값(`max`), 최소값(`min`)을 선택적으로 반환.
- 디폴트 파라미터 `metric="avg"`.
- 활용 예: 보고서용 요약 지표 자동 계산.

# 사용 예시

print(calc_metrics([10, 20, 30], "avg"))  # 20.0
print(calc_metrics([10, 20, 30], "max"))  # 30
print(calc_metrics([10, 20, 30]))         # 20.0 (디폴트 avg)
"""
```

```python
import numpy as np

def calc_metrics(data, metric="avg"):
  if data == []:
    return "값이 없으면 data가 없습니다."
  if metric == "avg":
    return np.mean(data)
  elif metric == "max":
    return np.max(data)
  elif metric == "min":
    return np.min(data)
```

```python
print(calc_metrics([10, 20, 30], "avg"))  # 20.0
print(calc_metrics([10, 20, 30], "max"))  # 30
print(calc_metrics([10, 20, 30], "min"))  # 10
print(calc_metrics([10, 20, 30]))         # 20.0 (디폴트 avg)
print(calc_metrics([]))                     # 값이 없으면 data가 없습니다
```

## 7. 이메일 제목 자동 필터링 함수(10점)

```python
"""
## 7. 이메일 제목 자동 필터링 함수

- 함수 이름: `filter_emails(subjects, keyword="광고")`
- 기능: 이메일 제목 리스트에서 특정 키워드 포함 여부를 조건문으로 검사 후 필터링.
- 출력: 키워드가 포함된 제목만 반환.
- 활용 예: 스팸/광고 메일 자동 분류.
"""
```

```python
def filter_emails(subjects, keyword="광고"):
  return_list = []
  for i in subjects:
    if keyword in i:
      return_list.append(i)
  return return_list

```

```python
emails = ["[광고] 특가 세일", "회의 일정 공유", "[광고] 쿠폰 발급", "업무 보고", "[일정]회의"]
keyword = "광고"
filter_emails(emails)
```
